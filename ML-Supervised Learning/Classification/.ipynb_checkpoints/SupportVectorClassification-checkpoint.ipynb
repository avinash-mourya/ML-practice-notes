```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
from mpl_toolkits.mplot3d import Axes3D
```

```python
data = pd.read_csv('iris1.csv')
```

```python
data.corr()
```

```python
#ax = plt.gca(projection="3d")
plt.figure(figsize=(10,8))
sn.scatterplot('petal_length','petal_width',data=data,hue="class",legend='brief')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.scatter(data['petal_length'],data['petal_width'],data['sepal_length'])
```

```python
X = data.iloc[:,2:4].values
Y = data.iloc[:,4].values
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
```

```python
train_x,test_x,train_y,test_y = train_test_split(X,Y,test_size=0.25,random_state=5)
```

```python
sc = StandardScaler()
train_x = sc.fit_transform(train_x)
test_x = sc.fit_transform(test_x)
Le = LabelEncoder()
train_y = Le.fit_transform(train_y)
test_y = Le.fit_transform(test_y)
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel='rbf',random_state=5)
```

```python
svc.fit(train_x,train_y)
```

```python
pred_y = svc.predict(test_x)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
confusion_matrix(test_y,pred_y)
```

```python
print(classification_report(test_y,pred_y))
```

```python
accuracy_score(test_y,pred_y)
```

```python
from matplotlib.colors import ListedColormap
plt.figure(figsize=(8,6))
x_set,y_set = train_x,train_y
x1,x2 = np.meshgrid(np.arange(start = x_set[:,0].min()-1,stop=x_set[:,0].max()+1,step = 0.01),np.arange(start = x_set[:,1].min()-1,stop=x_set[:,1].max()+1,step = 0.01))
plt.contourf(x1,x2,svc.predict(np.array([x1.ravel(),x2.ravel()]).T).reshape(x1.shape),alpha=0.60,cmap = ListedColormap(("red",'blue','green')))
plt.xlim(x1.min(),x1.max())
plt.ylim(x2.min(),x2.max())
for i,j in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set == j,0],x_set[y_set == j,1],c = ListedColormap(("red",'blue','green'))(i),label=j)
plt.title('Support Vector Classifier (Training set)')
plt.legend()
plt.show()
# ['setosa'=0,'versicolor'=1,'virginica'=2]
```

```python
plt.figure(figsize=(8,6))
x_set,y_set = test_x,pred_y
x1,x2 = np.meshgrid(np.arange(start = x_set[:,0].min()-1,stop=x_set[:,0].max()+1,step = 0.01),np.arange(start = x_set[:,1].min()-1,stop=x_set[:,1].max()+1,step = 0.01))
plt.contourf(x1,x2,svc.predict(np.array([x1.ravel(),x2.ravel()]).T).reshape(x1.shape),alpha=0.60,cmap = ListedColormap(("red",'blue','green')))
plt.xlim(x1.min(),x1.max())
plt.ylim(x2.min(),x2.max())
for i,j in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set == j,0],x_set[y_set == j,1],c = ListedColormap(("red",'blue','green'))(i),label=j)
plt.title('Support Vector Classifier (Testing set)')
plt.legend()
plt.show()
```
