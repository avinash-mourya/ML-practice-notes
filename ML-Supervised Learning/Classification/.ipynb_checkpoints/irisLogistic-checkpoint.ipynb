```python
X=data.iloc[:,0:4].values
Y=data.iloc[:,4].values
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.decomposition import PCA,KernelPCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
```

```python
train_x,test_x,train_y,test_y = train_test_split(X,Y,test_size=0.3,random_state=5,stratify=Y )
test_x[:,0]
```

```python
from sklearn.preprocessing import StandardScaler,LabelEncoder
sc = StandardScaler()
test_x = sc.fit_transform(test_x)
train_x = sc.fit_transform(train_x)
test_x
```

```python
LE = LabelEncoder()
train_y = LE.fit_transform(train_y)
test_y = LE.fit_transform(test_y)
```

### Linear Discreminant Analysis

```python
lda = LDA(n_components=2)
train_x = lda.fit_transform(train_x,train_y)
test_x = lda.fit_transform(test_x,test_y)
```

```python
LGS = LogisticRegression(random_state=10,solver='lbfgs', multi_class='auto')
LGS.fit(train_x,train_y)
```

```python
pred_y = LGS.predict(test_x)
```

```python
pred_y
```

```python
test_y
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
accuracy_score(test_y,pred_y)
```

```python
print(classification_report(test_y,pred_y))
```

```python
mat = confusion_matrix(test_y,pred_y)
mat
```

```python
# tick =['versicolor', 'virginica','setosa']
```

```python
# plt.figure(figsize=(8,8))
# sn.heatmap(c_m,  xticklabels=tick,yticklabels=tick,cmap='Blues_r',annot=True,square=True,robust=True, vmin=0,vmax=15)
# plt.show()
```

```python
from mlxtend.plotting import plot_confusion_matrix,plot_decision_regions,plot_pca_correlation_graph,plot_sequential_feature_selection
plot_confusion_matrix(mat,figsize=(3,3),show_absolute=False,show_normed =True)
plt.show()
#class_names=['setosa'=0,'versicolor'=1,'virginica'=2]
```

```python
from matplotlib.colors import ListedColormap
plt.figure(figsize=(8,6))
x_set,y_set = train_x,train_y
x1,x2 = np.meshgrid(np.arange(start = x_set[:,0].min()-1,stop=x_set[:,0].max()+1,step = 0.01),np.arange(start = x_set[:,1].min()-1,stop=x_set[:,1].max()+1,step = 0.01))
plt.contourf(x1,x2,LGS.predict(np.array([x1.ravel(),x2.ravel()]).T).reshape(x1.shape),alpha=0.60,cmap = ListedColormap(("red",'blue','green')))
plt.xlim(x1.min(),x1.max())
plt.ylim(x2.min(),x2.max())
for i,j in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set == j,0],x_set[y_set == j,1],c = ListedColormap(("red",'blue','green'))(i),label=j)
plt.title('Logistic Regretion (Training set)')
plt.legend()
plt.show()
# ['setosa'=0,'versicolor'=1,'virginica'=2]
```

```python
plt.figure(figsize=(8,6))
x_set,y_set = test_x,pred_y
x1,x2 = np.meshgrid(np.arange(start = x_set[:,0].min()-1,stop=x_set[:,0].max()+1,step = 0.01),np.arange(start = x_set[:,1].min()-1,stop=x_set[:,1].max()+1,step = 0.01))
plt.contourf(x1,x2,LGS.predict(np.array([x1.ravel(),x2.ravel()]).T).reshape(x1.shape),alpha=0.60,cmap = ListedColormap(("red",'blue','green')))
plt.xlim(x1.min(),x1.max())
plt.ylim(x2.min(),x2.max())
for i,j in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set == j,0],x_set[y_set == j,1],c = ListedColormap(("red",'blue','green'))(i),label=j)
plt.title('Logistic Regretion (Testing set)')
plt.legend()
plt.show()
```
