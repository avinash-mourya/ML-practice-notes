```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets 
from sklearn.linear_model import LinearRegression
```

```python
datas = datasets.load_diabetes
X = datas().data[:,None,8]
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_x,test_x,train_y,test_y = train_test_split(X,datas().target,test_size=0.3,random_state=42)
```

```python
train_x
```

```python
LR = LinearRegression()
LR.fit(train_x,train_y)
```

```python
result = LR.predict(test_x)
result
```

```python
test_x
```

```python
LR.coef_
```

```python
LR.intercept_
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
error = mean_squared_error(test_y,result)
```

```python
error
```

```python
score = r2_score(test_y,result)
score
```
