```python
X = data.iloc[:,0:13].values
Y = data.iloc[:,13].values
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor,plot_tree
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score,mean_squared_error
```

```python
train_x,test_x,train_y,test_y = train_test_split(X,Y,test_size=0.3,random_state=5)
```

```python
# DCR = KNeighborsRegressor()
```

```python
DCR = DecisionTreeRegressor(random_state=5)
```

```python
DCR.fit(train_x,train_y)
```

```python
pred_y = DCR.predict(test_x)
```

```python
test_y
```

```python
pred_y
```

```python
r2_score(test_y,pred_y)
```

```python
mean_squared_error(test_y,pred_y)
```

```python
plt.figure(figsize=(30,30),facecolor='m',edgecolor="red")
plot_tree(DCR.fit(train_x,train_y),rotate=True,max_depth=5,rounded=True,fontsize=20,impurity=True,node_ids=False, proportion=False,precision=2,filled=True)
plt.show()
```

# Ann - artificial neural network

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
```

```python
model.add(Dense(units=13,activation='relu',input_dim=13))
model.add(Dense(128,activation='relu'))
model.add(Dense(62,activation='relu'))
model.add(Dense(1,activation='linear'))
```

```python
model.compile(optimizer='adam',loss='mean_squared_error',metrics=['mse'])
```

```python
model.summary()
```

```python
model.fit(train_x,train_y,batch_size=100,epochs=150)
```

```python
pred_Y = model.predict(test_x)
```

```python
y = pred_Y.ravel()
```

```python
model.evaluate(test_x,test_y,batch_size=15)
```
