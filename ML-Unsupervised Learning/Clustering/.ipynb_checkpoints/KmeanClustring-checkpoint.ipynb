```python
X = data.iloc[:,0:8].values
```

```python
from sklearn.cluster import KMeans
```

# For measuring Number of cluster

```python
dst = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(X)
    dst.append(kmeans.inertia_)
```

```python
x = list(range(1,11))
dst/np.mean(dst)
```

Elbow plot use measuring a number of cluster

```python
plt.figure(figsize=(8,6))
plt.plot(x,dst)
plt.title('Elbow plot',fontdict = {'fontsize':30})
plt.xlabel('No. of cluster')
plt.ylabel('sqaured of distance')
plt.show()
```

3 cluster in this dataset

```python
Kmeans = KMeans(n_clusters=3,random_state=0)
p_k_y = Kmeans.fit_predict(X)
```

```python
p_k_y
```

```python
Kmeans.cluster_centers_
```

```python
color = ['red','green','blue','m','cyan']
```

```python
for i in np.unique(p_k_y):
    plt.scatter(X[p_k_y == i,3],X[p_k_y == i,4],c=color[i],label=f"cluster{i}")
plt.scatter(Kmeans.cluster_centers_[:,3],Kmeans.cluster_centers_[:,4],c='y')
plt.legend()
plt.show()
```
